```python
###Initialize, define X and Y
import numpy as np

x = np.asarray(new_dataframe.iloc[:, [0,1,2,3,4,5,7,8,9,10,11]])
x[0:5]
```

```python
y = np.asarray(new_dataframe['No Violations Observed'])
y [0:5]
print(y[0:5])
print("Initilaizing X and y complete")
```

```python
###Normalize data set

from sklearn import preprocessing
x = preprocessing.StandardScaler().fit(x).transform(x)
x[0:10]
```

```python
import lazypredict
#from lazypredict.Supervised
import LazyClassifier
#from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
#data = load_breast_cancer()
X = new_dataframe.data
y= new_dataframe.target
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=.5,random_state =123)
clf = LazyClassifier(verbose=0,ignore_warnings=True, custom_metric=None)
models,predictions = clf.fit(X_train, X_test, y_train, y_test)
models
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=4)
print ('Train set:', x_train.shape,  y_train.shape)
print ('Test set:', x_test.shape,  y_test.shape)
```

```python
from sklearn import svm

#Create a svm Classifier
clf = svm.SVC(kernel='linear') # Linear Kernel

#Train the model using the training sets
clf.fit(x_train, y_train)

#Predict the response for test dataset
y_pred = clf.predict(x_test)
print(y_pred)
```

```python
from sklearn import metrics

# Model Accuracy: how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
# Model Precision: what percentage of positive tuples are labeled as such?
print("Precision:",metrics.precision_score(y_test, y_pred, average='weighted'))

# Model Recall: what percentage of positive tuples are labelled as such?
print("Recall:",metrics.recall_score(y_test, y_pred, average='weighted'))
```

```python
# N Nearest Neighbor
# Find optimal K
from sklearn.neighbors import KNeighborsClassifier
#from sklearn.metrics import accuracy_score

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=4)

k_values = [i for i in range(1, 50, 2)]


k_acc_scores = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k, n_jobs = -1)
    cv_scores = cross_val_score(knn, x_train, y_train, cv=5, scoring='accuracy')
    k_acc_scores.append(cv_scores.mean())
    
optimal_k = k_values[k_acc_scores.index(max(k_acc_scores))]

print ("Our optimal k value is {}".format(optimal_k))


```

```python
from sklearn import neighbors
k = 5
cls = neighbors.KNeighborsClassifier(n_neighbors=k)
cls.fit(x_train, y_train)
accuracy =cls.score(x_train, y_train)
k_hat_pred = cls.predict(x_test)

print("Accouracy is {}". format(accuracy))
print(k_hat_pred)
```
